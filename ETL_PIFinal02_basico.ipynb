```python
df5 = pd.read_csv('./datastreams/05_accesos_a_internet_fijo_por_velocidad_bajada__y_provincia.csv')
df5
```

```python
df5.info()
```

```python
df5[df5.Año.isna()]
# se comprueba que hay un registros con valores nulos
```

```python
df5 = df5.dropna(subset=['Año', 'Trimestre', 'Provincia'], how= 'all')
# se eliminan todos los valores nulos en las columnas año, trimestre y provincia
```

```python
df5.info()
```

```python
df5.Año = df5.Año.astype(int)
df5.Trimestre = df5.Trimestre.astype(int)
```

```python
df5 = df5.applymap(lambda x: 0 if str(x) == '- 0' else x)
# cambio todos los ceros a numero
```

```python
columnas_flotantes = df5.columns.difference(['Año', 'Trimestre', 'Provincia'])
df5[columnas_flotantes] = df5[columnas_flotantes].applymap(lambda x: str(x).replace(',' , ''))
df5[columnas_flotantes] = df5[columnas_flotantes].applymap(lambda x: str(x).replace('.' , ''))
df5[columnas_flotantes] = df5[columnas_flotantes].astype(int)
```

```python
df5.head()
```

```python
df5.to_csv('./accesos_a_internet_fijo_por_velocidad_bajada__y_provincia.csv', index=False)
```
